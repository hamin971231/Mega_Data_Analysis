# 입지선정요인에 대한 가중치부여 기법에 관한 연구


- 정보계측기법의 경우 설문조사를 실시하지 않아도 되고, 의사결정 참여자의 주관적 판단이 아닌 정보라는 객관적 요소를 토대로 결과를 도출, 의사결정 타당성을 높이기 위해 정보수집계측의 범위를 넓히는 방법이 제안되어야한다. + 효율성을 높이기 위해 정보계측기법을 수행하기 위한 소프트웨어 마련되어야함
- 계측분석기법(AHP)의 의사결정 참여자 그룹 선정에 대한 합리성을 높이기 위해 의사결정 참여자 그룹 선정에 대한 객관적 기준이 확립되어야한다.




## 가중치 부여기법의 검토 및 이론적 고찰

### < 의사결정 기법의 검토 >
1) 평점모형 
- 분석자가 각 요인에 대한 상대적인 가중치를 부여하고 각 대안에 해당하는 값을 평점화 하는 방법이다.
- 전문가들의 주관적 의견에 의존한 점수부여 방식으로인해 설문 응답과정에서 전문가의 비일관성이 발생할 수 있음

2) 목표 달성평가법 
- 달성해야할 목표기준을 명시적으로 구분하고 각각의 목표별 성취점수를 산정 후 이를 종합
- 일반적으로 전문가의 지식과 경험을 바탕으로 가중치 결정
- 가중치 산정의 객관성 면에서 지나치게 전문가의 주관에 의존할 수 잇음

3) 다속성효용함수법 
- 의사결정을 위해 다양한 효용을 사용하는 방법 ( 효용: 의사결정에 미치는 요인들에 개인들이 부여하는 일종의 총체적 가치를 일컬음) 
- 요인들 간 상대적 중요도를 구할 수 있다는 장점
- 응답자들의 주관이 개입 + 효용함수 도출이 복잡하고 일관성 검증이 힘듦

4) 계층분석기법(AHP)
- 요소들간의 쌍대비교를 통한 판단행렬로부터 고유백터를 산출하여 중요도를 평가. 가장 많이 쓰임) 
- 집단 의사결정 방법 제공 
- 계층구조 형성에 대한 이론적 기초 부족 

## 계층분석기법의 (AHP)

- 계층분석기법은 의사결정의 목표나 검정기준이 다수이며, 복잡한 경우 상호 관련성이 적은 배타ㅓㄱ 대안들을 체계적으로 검정 할 수 있는 기법이다. 
- 기준들을 계층적으로 구성하여 비교 -> 기준들의 중요도를 파악할 수 잇다. 
- 동일한 단계에 있는 요소들 사이의 중요도를 관측하는 방법

### 계층분석기법과 정보계측기법의 비교

- AHP는 의견종합을 위한 수단으로 설문조사를 실시하는 반면에 정보계측기법의 경우는 의사결정 근거를 의사결정 대상의 정보에 두고 정보의 종합수단으로 총정보량의 산출값을 이용한다. 

- 의사결정을 위한 요인 선정 -> 의사결정 참여자 그룹 선정 -> 참여자들의 의견 종합을 위한 설문 -> 일관성 검증 -> 의사결정 도출
- 정보계측기법 : 요인선중 후 요인에 대한 정보수집 -> 정보내에서 명사를 추출

## 계층분석기법
1. 전문가 설문조사 
- 입지선정 요인을 바탕으로 설문지 작성(5점척도)
2. 일관성 검증 
- 설문을 통해 구해진 각 요인들의 상대적인 중요도는 일관성이 있는지 검증해야함.
- 정량적 분석에 의해 결과값을 쌍대비교, 기준이나 대안의 수가 적당해야함.
- 일관성 비율로 검증하여, 일관성 비율이 0.1이하의 값을 가질떄 일관성이 있는 것으로 판단.
