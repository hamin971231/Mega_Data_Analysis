```python
def multipolygon_to_coordinates(x):
    lon, lat = x[0].exterior.xy
    return [[x, y] for x, y in zip(lon, lat)]

## 다각형에서 평행좌표
def polygon_to_coordinates(x):
    lon, lat = x.exterior.xy
    return [[x, y] for x, y in zip(lon, lat)]

emer_bell = gpd.GeoDataFrame(df,geometry=gpd.points_from_xy(df['WGS84경도'],df['WGS84위도']))
emer_bell.set_crs(epsg = 4326,inplace=True)
```

```python
import re
import pydeck as pdk
# 지도 시각화
emer_bell = emer_bell[['WGS84위도','WGS84경도']]  # '총생활인구수'를 시각화에 추가, 필요에 따라 수정
layer1 = pdk.Layer(
    'ScatterplotLayer',
    emer_bell,
    get_position='[WGS84경도, WGS84위도]',
    get_radius=7,
    get_fill_color='[255, 0, 0]',
    pickable=True,
    auto_highlight=True
)

center = [127.0040, 37.5604]  # 중구의 중심 좌표로 변경
view_state = pdk.ViewState(
    longitude=center[0],
    latitude=center[1],
    zoom=13
)

r = pdk.Deck(layers= layer1, initial_view_state=view_state)
r.to_html('./wow.html')
```
