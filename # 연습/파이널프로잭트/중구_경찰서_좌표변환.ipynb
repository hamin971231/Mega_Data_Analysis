```python
police.loc[police['관서명']=='약수','주소']='서울특별시 중구  동호로 5길 15'
police.loc[police['관서명']=='신당','주소']='서울특별시 중구 다산로 248'

police.loc[police['관서명']=='명동','주소']='서울특별시 중구 명동길 30'

```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')
```

```python
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y=[geo.latitude,geo.longitude]
        return x_y
    except :
        return [0,0]
    
```

```python
address = police['주소']
```

```python
latitude = []
longitude=[]

for i in address :
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])

```
