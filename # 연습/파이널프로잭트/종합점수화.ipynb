```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import numpy as np 
```

```python
data = pd.read_excel('중구_최종데이터.xlsx')
data.head()
```

```python
data.columns
```

```python
data = data.loc[:,'노드링크 유형':]
data.head()
```

```python
need = data.loc[:,'CCTV_100m':]
need.head()
```

```python
scaler = MinMaxScaler()
need_std = scaler.fit_transform(need)
need_std_data = pd.DataFrame(need_std,columns=need.columns)
need_std_data
```

```python
data['CCTV_100m_std'] = need_std_data['CCTV_100m']
data['숙박업소_100m_std'] = need_std_data['숙박업소_100m']
data['유흥업소_100m_std'] = need_std_data['유흥업소_100m']
data['경찰서_500m_std'] = need_std_data['경찰서_500m']
data['비아파트_25m_std'] = need_std_data['비아파트_25m']
data['취약계층_평균(생활인구)_std'] = need_std_data['취약계층_평균(생활인구)']
data.head()
```

```python
data.columns
```

```python
score = data.loc[:,'CCTV_100m_std':]

score.head()
```

```python
score['범죄종합점수'] = -0.074 * score['CCTV_100m_std'] + 0.065 * score['숙박업소_100m_std'] + 0.23 * score['유흥업소_100m_std'] - 0.23 * score['경찰서_500m_std'] + 0.259 * score['비아파트_25m_std'] + 0.169 * score['취약계층_평균(생활인구)_std'] 
score['범죄종합점수']
```

```python
score.head()
```

```python
data['범죄종합점수'] = score['범죄종합점수']
data.head()
```

```python
data.to_excel('중구_후보지+범죄종합점수.xlsx')
```
