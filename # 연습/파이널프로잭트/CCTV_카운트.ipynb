```python
### 노드데이터 좌표계설정

node = node[(node['노드링크 유형'] == 'NODE') &
        (node['육교'] == 0) &
        (node['횡단보도'] == 0)]

def multipolygon_to_coordinates(x):
    lon, lat = x[0].exterior.xy
    return [[x, y] for x, y in zip(lon, lat)]

## 다각형에서 평행좌표
def polygon_to_coordinates(x):
    lon, lat = x.exterior.xy
    return [[x, y] for x, y in zip(lon, lat)]

## 지리데이터프레임으로 변환

import geopandas as gpd
gs = gpd.GeoSeries.from_wkt(node['노드 WKT'])
## 좌표계 설정하는 거 같음
# df_dobo_100 = gpd.GeoDataFrame(df, geometry = gs, crs = 'epsg:4326')
node_gs = gpd.GeoDataFrame(node, geometry = gs, crs = 'epsg:4326')
node_gs.head()
```

```python
## CCTV 데이터 지리데이터로 ㄱㄱ 

cctv = gpd.GeoDataFrame(cctv,geometry=gpd.points_from_xy(cctv['WGS84경도'],cctv['WGS84위도']),crs = 'epsg:4326')
cctv.head()
```

```python
import math
cctv_locations = [(37.5654, 126.9891), (37.5618, 126.9775), (37.5667, 126.9816)]
center = (127.00138520146373,37.568187437157185)
radius = 1

def count_cctv_within_radius(cctv_locations, center, radius):
    count = 0
    for cctv in cctv_locations:
        x, y = cctv
        distance = math.sqrt((x - center[0])**2 + (y - center[1])**2)
        if distance <= radius:
            count += 1
    return count
cctv_count = count_cctv_within_radius(cctv_locations, center, radius)
print(f"반경 {radius} 내에 있는 CCTV 개수: {cctv_count} 개")
```
