```python
from geopy.distance import geodesic

x= node_gs['geometry'].x
y =node_gs['geometry'].y
node_data2 = [(y,x) for y,x in zip(y,x)]

bell_x= bell['geometry'].x
bell_y =bell['geometry'].y
bell_loc_data = [(y,x) for y,x in zip(bell_y,bell_x)]
bell_loc_data
```

```python
len(bell_loc_data)
```

```python

radius = 30


# 노드 좌표별로 반경 내에 있는 CCTV 개수 계산
result_list = []

for node_coordinate in node_data2:
    node_count = 0
    for bell in bell_loc_data:
        distance = geodesic(node_coordinate, bell).meters
        if distance <= radius:
            node_count +=1
    
    result_list.append(node_count)
        
result_list
```

```python
node_gs['수요충족여부'] = result_list
```
