```python
from geopy.distance import geodesic

x= node_gs['geometry'].x
y =node_gs['geometry'].y
node_data2 = [(y,x) for y,x in zip(y,x)]

hotel_x= hotel['geometry'].x
hotel_y =hotel['geometry'].y
hotel_loc_data = [(y,x) for y,x in zip(hotel_y,hotel_x)]
hotel_loc_data
from geopy.distance import geodesic

# 일단 중구 중심 좌료 ( 경도 , 위도 )

## CCTV 좌표데이터 (예시 , 경도 , 위도 )
## cctv_loc_data

## 중구 내의 모든 노드 좌표 데이터 예시

# node_data

# 반경 (미터)

radius = 100

# 노드 좌표별로 반경 내에 있는 CCTV 개수 계산
```

```python


# 노드 좌표별로 반경 내에 있는 CCTV 개수 계산
hotel_counts_for_nodes = []

for node_coordinates in node_data2:
    hotel_count = 0
    for hotel_location in hotel_loc_data:
        distance = geodesic(node_coordinates, hotel_location).meters
        if distance <= radius:
            hotel_count += 1
    hotel_counts_for_nodes.append(hotel_count)

# 결과 출력
for i, node_coordinates in enumerate(node_data2):
    print(f"Node {i + 1}: 좌표 {node_coordinates}, {radius}미터 반경 내에 있는 숙박업소 개수: {hotel_counts_for_nodes[i]} 개")
    
```

```python
node['숙박업소_100m'] = hotel_counts_for_nodes
node
```
