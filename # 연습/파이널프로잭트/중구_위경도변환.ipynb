```python
import pandas as pd
import numpy as np
import pyproj
import folium
```

# 관악구_단란주점


```python
df = pd.read_csv('중구_단란주점.csv',encoding='cp949')
df.head()
```

```python
df['좌표정보(X)'] = pd.to_numeric(df['좌표정보(X)'],errors='coerce')
df['좌표정보(Y)'] = pd.to_numeric(df['좌표정보(Y)'],errors='coerce')
```

```python
def project_array(coord, p1_type, p2_type):
    p1 = pyproj.Proj(init=p1_type)
    p2 = pyproj.Proj(init=p2_type)
    fx, fy = pyproj.transform(p1, p2, coord[:, 0], coord[:, 1])
    return np.dstack([fx, fy])[0]
    
```

```python
coord = np.array(df[['좌표정보(X)','좌표정보(Y)']])
coord
```

```python
p1_type = "epsg:2097"
p2_type = "epsg:4326"
result = project_array(coord, p1_type, p2_type)
result
```

```python
result=pd.DataFrame(result,columns=['경도','위도'])
result
```

```python
danran = pd.read_excel('관악구_단란주점.xlsx')
danran['경도']=result['경도']
danran['위도']=result['위도']
danran.to_excel('관악구_단란주점.xlsx')
```
