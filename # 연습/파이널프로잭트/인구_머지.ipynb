```python
import pandas as pd

node = pd.read_excel('중구_노드_링크_비아파트추가.xlsx')
pop = pd.read_excel('중구_취약계층_집계(최종).xlsx')
node.head()
```

```python
pop.head()
```

```python
final_data = pd.merge(node,pop[['평균','행정동코드']],how='left',right_on='행정동코드',left_on='행정동코드')
final_data.rename(columns={'평균':'취약계층_평균(생활인구)'},inplace=True)
final_data.head()
```

```python
final_data.to_excel('중구_최종데이터.xlsx')
```
