```python
train = pd.read_excel('범죄.xlsx', index_col='Unnamed: 0')
train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train = scaler.fit_transform(train)
train = pd.DataFrame(train)
train.columns = ['전국', '아파트/연립다세대','단독주택', '노상', '유흥숙박업소', '유흥업소', '상점', '사무실', '피씨방',
       '고속도로', '백화점', '슈퍼마켓', '편의점', '공장', '지하철', '유원지', '학교', '의료기관', '종교기관',
       '주차장', '화장실', '창고', '역, 대합실', '기타 교통수단', '흥행장', '금융기관', '산야']
train.columns
```

```python
names = train.columns
n = len(names)
pv = 0.05

data = []

for i in range(0, n):
    fields = names[1] + 'vs' + names[i]
    s, p = stats.pearsonr(train[names[1]], train[names[i]])
    result = p < pv

    data.append({'fields': fields, 'statistic':s, 'pvalue': p, 'result': result})

rdf = DataFrame(data)
rdf.set_index('fields', inplace=True)
rdf
```

```python
x = train.loc[:, ['아파트/연립다세대', '노상', '유흥숙박업소', '유흥업소','상점', '사무실', '피씨방',
       '고속도로', '백화점', '슈퍼마켓', '편의점', '공장', '지하철', '유원지', '학교', '의료기관', '종교기관',
       '주차장', '화장실','창고', '역, 대합실', '기타 교통수단', '흥행장', '금융기관', '산야']].values
x
```

```python
import statsmodels.api as sm
results = sm.OLS(train['전국'], sm.add_constant(x)).fit()
print(results.summary())
```

```python
import statsmodels.api as sm
model = sm.OLS(train['전국'], train[['아파트/연립다세대', '유흥숙박업소', '학교','슈퍼마켓','화장실']])
fit = model.fit()
tbl = fit.summary()
```

```python
import re

my = {}

for k in range(0, 3, 2):
    items = tbl.tables[k].data
    #print(items)

    for item in items:
        #print(item)
        n = len(item)

        for i in range(0, n, 2):
            key = item[i].strip()[:-1]
            value = item[i+1].strip()

            if key and value:
                my[key] = value

my
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
for i in range(0, len(model.exog_names)):
    vif = variance_inflation_factor(model.exog, i)

    if vif < 10:
        print("%s의 VIF: %f (good)" % (model.exog_names[i], vif))
    else:
        print("%s의 VIF: %f (bad)" % (model.exog_names[i], vif))
```

```python
my['variables'] = []

for i, v in enumerate(tbl.tables[1].data):
    if i == 0:
        continue

    # 변수의 이름
    name = v[0].strip()
    # 변수의 이름 목록
    name_list = list(model.exog_names)
    # 변수의 이름 목록에서 현재 변수가 몇 번째 항목인지 찾기 
    j = name_list.index(name)

    vif = 0

    # 0번째인 Intercept는 제외
    if j >= 0:
        vif = variance_inflation_factor(model.exog, j)

    my['variables'].append({
        "name": name,
        "coef": v[1].strip(),
        "std err": v[2].strip(),
        "t": v[3].strip(),
        "P-value": v[4].strip(),
        "Beta": 0,
        "VIF": vif,
    })

my
```

```python
mylist = []

for i in my['variables']:
    if i['name'] == 'Intercept':
        continue

    item = {
        "독립변수": i['name'],
        "B": i['coef'],
        "표준오차": i['std err'],
        "β": i['Beta'],
        "t": "%s*" % i['t'],
        "유의확률" : i['P-value'],
        "VIF": i["VIF"]
    }

    mylist.append(item)

df = DataFrame(mylist, index=['전국','전국','전국','전국','전국'])
df.index.name = '종속변수'
df
```
