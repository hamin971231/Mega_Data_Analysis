```python
for i,v in enumerate(bad.loc[bad['도로명주소'].isnull(),'지번주소'].str.split(' ')) :
    bad.loc[bad['도로명주소'].isnull(),'시'] = v[0]
    bad.loc[bad['도로명주소'].isnull(),'구'] = v[1]
    bad.loc[bad['도로명주소'].isnull(),'상세주소'] = v[2]
    # print(v[2])


bad.loc[bad['도로명주소'].isnull(),:]
    
    
```

```python
bad['구'].unique()
```

```python
test = bad.groupby(['구']).count()
test
```

```python
test['유흥업소개수']=test['Unnamed: 0']
유흥업소_구단위_현황 = test.iloc[:,[-1]]
유흥업소_구단위_현황
```
