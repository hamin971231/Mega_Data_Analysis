## #02. 데이터 가져오기

```python
digits = load_digits()
```

### 무작위로 50개 추출

#### 추출할 이미지 수, 랜덤시드 고정


```python
rnd_count = 50

np.random.seed(342)
```

### 50개에 대한 인덱스 생성

```python
idx= np.random.choice(range(len(digits.data)),rnd_count)
len(idx)
```

#### 무작위 리스트에 대한 표본추출

```python
x= digits.data[idx]
x.shape
```

```python
images = digits.images[idx]
images.shape
```

```python
fix ,ax = plt.subplots(5,10,figsize = (30,5))

idx = 0 
for i in range(0,5):
    for j in range(0,10):
        # ax[i][j]
        ax[i][j].imshow(images[idx],cmap='gray')
        ax[i][j].set_title(str(idx))
        idx+=1
```

## #03. 군집수행

```python
linked = linkage(x,'ward')
linked
```

```python
from matplotlib.offsetbox import OffsetImage,AnnotationBbox


plt.figure(figsize=(20,10))
ddata = dendrogram(linked)
plt.show()
plt.close()
```
